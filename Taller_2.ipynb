```python
import numpy as np
import pandas as pd
import wooldridge as wd
```

```python
wage = wd.data('wage1')[['wage','educ','exper']]
wage
```

```python
y=wage['wage'].to_numpy()
x=wage[['educ','exper']].to_numpy()
u = wage.to_numpy()
```

```python
x.T
```

```python
np.linalg.inv(x.T.dot(x))
```

```python
betas=np.linalg.inv(x.T.dot(x)).dot(x.T.dot(y))
betas
```

```python
u
```

```python
x.dot(betas)
```

```python
y_est=x.dot(betas)
```

```python
y-y_est
```

```python
u_est=y-y_est
```

```python
y.shape #número de datos totales
```

```python
sum(u_est**2)
```

```python
n,k=x.shape
n,k
```

```python
sigma=sum(u_est**2)/(n-k)
sigma
```

```python
x.T
```

```python
x.T.dot(x)
```

```python
Error_est=np.diag(np.sqrt(np.linalg.inv(x.T.dot(x))*sigma))
Error_est
```

```python
t= betas/Error_est
t
```

```python
int_confianza_sup= betas+1.96*(Error_est)
int_confianza_sup
```

```python
int_confianza_inf= betas-1.96*(Error_est)
int_confianza_inf
```

```python
# Los dos intervalos de confianza son estadísticamente significativos
```
